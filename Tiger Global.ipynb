### function: Def scrape_13F ##

```python
# write a function to scrape the files
def scrape_13f(file):
    html = open(file).read()
    soup = BeautifulSoup(html, "html.parser")
    table = soup.find_all('table')
    df = pd.read_html(str(table))[6]
    return df
```

### function: Def df_13F_Edit ###

```python
def df_13F_edit(df):    
    df.columns = ['Company', 'Title of Class', 'CUSIP', '$ Value', 'Shares', 'PRN', 'Call', 'Discretion', 'Manager',
             'Sole', 'Shared', 'None']
    df_raw = df[3:]
    return df_raw
```

### function: Def df_compessed ###

```python
def df_compressed(df_raw):
    # subset df 
    df = df_raw[['Company','CUSIP', '$ Value', 'Shares']]
    df['$ Value'].astype(int)
    df['Shares'].astype(int)
    df_ = df
    return df_ 
```

### function: Def port_calc ###

```python
def calc_portfolio(df_):
    df_['$ Value'] = df_['$ Value'].astype(int)
    df_['Shares'] = df_['Shares'].astype(int)  
    total = df_['$ Value'].sum()
    df_.loc[:,'% Portfolio'] = (df_['$ Value'])/(df_['$ Value'].sum())*100
    df_port = df_
    return df_port
```
